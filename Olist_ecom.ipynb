###  How much did sellers make from each product categories


```python

query1 = """
            SELECT (products.product_category_name) AS product_name, order_items.seller_id, order_items.price, order_items.freight_value
            FROM order_items JOIN products 
            ON order_items.product_id = products.product_id
            JOIN orders
            ON orders.order_id = order_items.order_id
            GROUP BY orders.order_id
            
"""
```

```python
### Freight Value: Transit insurance is taken by the Shipper to protect against any loss, damage or pilferage during transit of the shipment.
```

```python
freq_orders = mysql(query1)
freq_orders
```

```python
freq_orders['profit'] = freq_orders.price - freq_orders.freight_value
freq_orders
```

```python
freq_order_by_seller = pd.pivot_table(freq_orders, values=['profit'], index=['seller_id', 'product_name'], aggfunc=np.sum)
freq_order_by_seller
```
