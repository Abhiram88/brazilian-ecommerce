```python
# most sold products by seller

query = """SELECT (products.product_category_name) AS product_name, order_items.seller_id 
            FROM order_items JOIN products 
            ON products.product_id = order_items.product_id 
            GROUP BY products.product_id"""
```

```python
products_by_seller = mysql(query)
```

```python
# most profitable product for a seller

query1 = """
            SELECT (products.product_category_name) AS product_name, order_items.seller_id, order_items.price
            FROM order_items JOIN products 
            ON order_items.product_id = products.product_id
            JOIN orders
            ON orders.order_id = order_items.order_id
            GROUP BY orders.order_id
            
"""
```

```python
freq_orders = mysql(query1)
freq_orders
```

```python
freq_order_by_seller = pd.pivot_table(freq_orders, values=['price'], index=['seller_id', 'product_name'], aggfunc=np.sum)
freq_order_by_seller
```
